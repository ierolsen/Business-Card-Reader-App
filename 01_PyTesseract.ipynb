# Image to Data

```python
data = pytesseract.image_to_data(img_pil)
```

```python
print(data)
```

```python
data
```

```python
data.split("\n")
```

```python
list(map(lambda x: x.split("\t"), 
         ["'level\tpage_num\tblock_num\tpar_num\tline_num\tword_num\tleft\ttop\twidth\theight\tconf\ttext'"]))
```

```python
dataList = list(map(lambda x: x.split("\t"), data.split("\n")))
```

```python
dataList
```

```python
df = pd.DataFrame(dataList[1:], columns=dataList[0])

df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True) # it will remove the missing values in rows

col_int = ["level", "page_num", "block_num", "par_num", "line_num", "word_num", "left", "top", "width", "height", "conf"]
df[col_int] = df[col_int].astype(int)
```
